```python
import pandas as pd
import numpy as np
import plotly.express as px

Oil_visc = np.array([50,375,50,350,350,500,500,1000,100,500,500,1000,1000,1000,5000,20,20,50,100,100,100,500,20,100,350,500,1000,5000,5000,5000,375,50,20,50,500,500,375,375,5,10])
Standoff_nonD = np.array([1.8,1.9,1.9,0.8,1.0,0.8,1.0,1.0,2.1,1.9,2.2,1.1,1.3,2.6,1.9,2.3,2.1,1.1,1.4,1.3,1.9,1.4,1.1,1.7,1.6,1.4,1.5,1.2,1.2,1.6,1.5,1.8,1.7,1.2,2.2,1.4,1.1,1.4,1.5,1.3])
Oil_t = np.array([1.2,1.1,1.5,0.4,0.3,0.5,0.4,0.9,0.8,1.3,0.5,0.8,0.8,1.7,1.3,1.1,1.0,0.9,1.1,0.9,1.4,0.7,1.0,0.9,1.2,0.8,1.3,1.1,1.0,0.6,1.0,0.9,1.1,0.9,1.9,1.0,0.8,0.7,0.7,0.8])
Behaviour = 'W','G','W','W','G','G','G','G','W','G','G','G','G','G','G','W','W','W','W','W','W','G','W','W','W','C','N','N','N','G','W','W','C','W','G','G','W','W','C','C'

df = pd.DataFrame({
'cat':Behaviour, 'col_x':Oil_visc, 'col_y':Standoff_nonD, 'col_z':Oil_t
})
fig = px.scatter_3d(df, x='col_x', y='col_y', z='col_z',
                    color='cat',
                    title="Behaviour Map")
fig.update_layout(scene = dict(
                    xaxis_title='Oil viscosity (cSt)',
                    xaxis_type='log',
                    yaxis_title='Standoff/Rmax',
                    zaxis_title='Oil thickness/Rmax'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig.show()
```

```python
#read in from csv
import pandas as pd
import numpy as np
import plotly.express as px

fpath = '/content/drive/MyDrive/2021/Oil Layer Experiment/B_map1.csv'
df = pd.read_csv(fpath)
#df.head()
V1 = np.array(df.iloc[:,[1]]).flatten()
S1 = np.array(df.iloc[:,[4]]).flatten()
O1 = np.array(df.iloc[:,[6]]).flatten()
B1 = np.array(df.iloc[:,[7]]).flatten()

fpath2 = '/content/drive/MyDrive/2021/Oil Layer Experiment/B_map2.csv'
df2 = pd.read_csv(fpath2)
V2 = np.array(df2.iloc[:,[1]]).flatten()
S2 = np.array(df2.iloc[:,[4]]).flatten()
O2 = np.array(df2.iloc[:,[6]]).flatten()
B2 = np.array(df2.iloc[:,[7]]).flatten()

#add the data to the first set
Viscosity = np.append(V1, V2)
Standoff = np.append(S1, S2)
OilThickness = np.append(O1, O2)
Behaviour = np.append(B1, B2)

df = pd.DataFrame({
'cat':Behaviour, 'col_x':Viscosity, 'col_y':Standoff, 'col_z':OilThickness
})
fig = px.scatter_3d(df, x='col_x', y='col_y', z='col_z',
                    color='cat',
                    title="Behaviour Map")
fig.update_layout(scene = dict(
                    xaxis_title='Oil viscosity (cSt)',
                    xaxis_type='log',
                    yaxis_title='Standoff/Rmax',
                    zaxis_title='Oil thickness/Rmax'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig.show()
```

```python
fig = px.scatter(df, x="col_x", y="col_y", color="cat", log_x=True, range_x=[2,6000], range_y=[0.5,3.0],
                 labels=dict(col_x="Oil viscosity (cSt)", col_y="Standoff/Rmax"),
                 width=600, height=500)
fig.update_layout(plot_bgcolor='#e0e0e0')
fig.update_traces(marker={'size': 12})
fig.show()
```

```python
fig = px.scatter(df, x="col_x", y="col_z", color="cat", log_x=True, range_x=[2,6000], range_y=[0,2.0],
                 labels=dict(col_x="Oil viscosity (cSt)", col_z="Oil thickness/Rmax"),
                 width=600, height=500)
fig.update_layout(plot_bgcolor='#e0e0e0')
fig.update_traces(marker={'size': 12})
fig.show()
```
